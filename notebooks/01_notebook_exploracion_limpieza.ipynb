### Preparación y limpieza de datos

- En el caso de la variable PetID (Identificador del animal) no es necesaria para la predicción. Por lo que esa variable va a ser eliminada.  
- Creo que para una mejor interpretación y comprensión de los datos en la presentación, es necesario cambiar las edades de meses a años.  
- Para manipular mejor la variable target, voy a cambiarle el nombre de "AdoptionLikelihood" a "target"



```python
train, test = train_test_split(df_adopciones, test_size=0.2, stratify=df_adopciones['AdoptionLikelihood'], random_state=42)
```

```python
# Elimino PetID:

train = train.drop(columns = ["PetID"], axis = 1)
test = test.drop(columns = ["PetID"], axis = 1)
```

```python
# Transformación de edad de meses a años:

train["AgeYears"] = round(df_adopciones["AgeMonths"]/12, 2)
test["AgeYears"] = round(df_adopciones["AgeMonths"]/12, 2)
```

```python
# Elimino columna de AgeMonths

train = train.drop(columns = ["AgeMonths"], axis = 1)
test = test.drop(columns = ["AgeMonths"], axis = 1)
```

```python
# Cambio nombre de target

train = train.rename(columns = {"AdoptionLikelihood": "target"})
test = test.rename(columns = {"AdoptionLikelihood": "target"})
```

```python
train.to_csv('../data/Raw/train_data_raw.csv', index = False)
train.head()
```

##### Datos limpios:

- Transformación de edad de meses a años
- Cambio de nombre de la variable target

```python
features = train.columns.to_list()
```

```python
# Transformación de las variables:

train = pd.get_dummies(train[features], columns = ["PetType", "Breed", "Color", "Size"], dtype = int)
test = pd.get_dummies(test[features], columns = ["PetType", "Breed", "Color", "Size"], dtype = int)
```

```python
train.to_csv('../data/Clean/train_data_clean.csv', index = False)
test.to_csv('../data/Clean/test_data_clean.csv', index = False)
```
